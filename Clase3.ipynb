```python
notas = [2,4,6,9,10,3,5,6]
```

```python
sum(notas)
```

```python
suma = 0
for nota in notas:
  suma += nota             # al valor que ya tenia mi variable "suma", sumarle el valor que se encuentra en el elemento dentro del vector
```

```python
print(suma)
```

```python
print("El promedio es ",suma/len(notas))
```

```python
x = [23,54,67]
y = [90,76,34]
```

```python
plt.plot(x,y,c="green")
```

```python
# Puedo cambiar el tipo de línea
plt.plot(x,y,linestyle="dashed")
```

```python
# Marcar intersecciones y establecer su tamaño
plt.plot(x,y,marker="*", ms=15)
plt.show() # Elimina info sobre el guardado
```

```python
# Representar líneas en un mismo gráfico
y_1 = [3,8,1,10]
y_2 = [6,2,7,11]
plt.plot(y_1,marker="*") # Cuando no le otorgamos el otro parametro, la variable X, usa los numeros enteros y empieza del 0
plt.plot(y_2,linestyle="dashed",c="pink",marker="o")  # Concatenamos otro gráfico
plt.show()
```

```python
# Un ejemplo: cansancio vs pulsaciones
x = [60,70,85,90,100,120]
y = [10,20,40,50,70,80]
len(x),len(y)
```

```python
plt.plot(x,y,c="red",marker="o",ms=10)
# Agrego título
# Puedo localizarlo
plt.title("Cansancio vs pulsaciones",loc="left")
plt.xlabel("Pulsaciones")  # Nombro los ejes
plt.ylabel("Cansancio")
plt.grid() #Añado grilla al gráfico
plt.show() # Borro info innecesaria
```

```python
# Subplots
x
```

```python
pulsaciones = x
calorias = [20,40,50,60,75,90]
```

```python
plt.subplot(1,2,1) # Este gráfico tiene 1 fila, 2 columnas y el primero está en la columna 1
# Copio lo anterior
plt.plot(x,y,c="red",marker="o",ms=10)
plt.title("Cansancio vs pulsaciones",loc="left")
plt.xlabel("Pulsaciones")
plt.ylabel("Cansancio")
plt.grid()

plt.subplot(1,2,2)
plt.plot(pulsaciones,calorias,c="green")
plt.title("Pulsaciones vs calorias")
plt.xlabel("Pulsaciones")
plt.ylabel("Calorias")
plt.grid()

plt.suptitle("TP Intregrador") # Añado un "super título" a los subplots

# Así añado dos gráficos dentro de la misma fila
```
